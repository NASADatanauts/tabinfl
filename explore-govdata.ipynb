```r
library(jsonlite)
metadata <- fromJSON("https://catalog.data.gov/api/3/action/package_search?rows=100")
names(metadata$result$results)
sapply(metadata$result$results, class)

```

```r
# library(dplyr)

govresults <- metadata$result$results
govtitle <- data_frame(title = govresults$title,
                       notes = govresults$notes,
                       num_res = govresults$num_resources,
                       owner = govresults$organization$title,
                        id = govresults$id
                       )
govtitle
```
